## Review samples count for each class
### Fs = 2500    
### Rest = 4s ( 4 x 2500 = almost 1000 samples)
### Trial = 3s (Visual Cue) + 4s (Movement Execution RM / MI)
### Trial = 3 x 2500 + 4 x 2500 = almost 17500 samples
<img src="Slide1.JPG" style="width:800px;hight:300px;"/>

```python
start_point_info = data['mrk'][0][0][5][0][0][0][0][0] 
end_point_info = data['mrk'][0][0][5][0][0][0][0][1]
trigger_marking_time = data['mrk'][0][0][0][0]
trigger_number = data['mrk'][0][0][1][0]
label_decode ={11:'Forward' , 21:'Backward'  , 31:'Left' , 41:'Right' , 51:'Up' , 61:'Down' , 8:'Rest'}
Forward,Backward,Left,Right,Up,Down,Rest=[],[],[],[],[],[],[]

for i in range (0,data['nfo'][0][0][8][0][0]): # data['nfo'][0][0][8][0][0] --> 600 (300 trial + 300 rest period)

        start,end=0,0

        if i==0:
            start=start_point_info
            end=trigger_marking_time[i]
        elif i==599:
            start=trigger_marking_time[i]
            end=end_point_info
        else:
            start=trigger_marking_time[i-1]
            end=trigger_marking_time[i]

        if label_decode[trigger_number[i]]=='Forward':
             Forward.append(end-start)
        elif label_decode[trigger_number[i]]=='Backward':
             Backward.append(end-start)
        elif label_decode[trigger_number[i]]=='Left':
             Left.append(end-start)
        elif label_decode[trigger_number[i]]=='Right':
             Right.append(end-start)
        elif label_decode[trigger_number[i]]=='Up':
             Up.append(end-start)
        elif label_decode[trigger_number[i]]=='Down':
             Down.append(end-start)
        else:
             Rest.append(end-start)
        
```

```python
print('Rest      ',min(Rest),'      ',max(Rest))
print('Forward   ',min(Forward),'     ',max(Forward))
print('Backward  ',min(Backward),'     ',max(Backward))
print('Left      ',min(Left),'     ',max(Left))
print('Right     ',min(Right),'     ',max(Right))
print('Up        ',min(Up),'     ',max(Up))
print('Down      ',min(Down),'     ',max(Down))
```

```python
def fun(list,threshold):
    for i in list:
        if i > threshold : 
            print(i)
```
